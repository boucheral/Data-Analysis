```python
emailPattern = re.compile('.+@enron.com')

def skip(a,n):
    
    if emailPattern.search(n) :
        return True
    
    if a.startswith('enron'):
        return True
    
    if '<' in a :
        return True
    
    bad = ['enron', '_', 'employees', 'team', 'temp', 'administrator']
    
    for x in bad :
        if x in n.lower() :
            return True
        
    return False
    
def swap(s) :
    if (',' in s) :
        x = s.split(',')
        return x[1].strip() + ' ' + x[0].strip()
    else :
        return s
    
def emailsAndNames(address, names) :
    
    regexp1 = re.compile('<[^>]+>|.+@.+')
    re2 = re.compile('<[^>]+>[,]?')
    
    if (address is None or names is None) :
        return None
    
    a = [x.strip() for x in address.split(',')]

    if (re2.search(names)) :
        n = re2.sub('|', names )
        n = [swap(x.strip()) for x in n.split('|')]
    else :
        n = [x.strip() for x in names.split(',')]
   
    d = {}
    x = 0
    while x < len(a) and x < len(n) :
        if skip(a[x], n[x]) :
            x = x + 1
            continue
        if emailPattern.match(a[x]) :
            d[a[x]] = n[x]
        x = x + 1
        
    if d :
        return d
    else :
        return None

def getAddresses(email) :
    to = email['to']
    tox = email['X-To']
    f = email['from']
    fx = email['X-From']
    cc = email['cc']
    ccx = email['X-Cc']
    
    addresses = {'from' : emailsAndNames(f, fx), 'to' : emailsAndNames(to, tox), 'cc' : emailsAndNames(cc, ccx)} 
    
    if (not addresses['from'] and not addresses['to'] and not addresses['cc']):
        return None
    else :
        return addresses
    
addresses = """l..garcia@enron.com, michael.seely@enron.com, 
    carrie.larkworthy@enron.com,john.kinser@enron.com, 
    joe.errigo@enron.com,russell.ballato@enron.com, jason.choate@enron.com,
	patrick.hanse@enron.com, lisa.burnett@enron.com,
	e-mail <.david@enron.com>, e-mail <.jeff@enron.com>,
	e-mail <.joe@enron.com>, e-mail <.joe@enron.com>,
	e-mail <.kari@enron.com>, e-mail <.lalena@enron.com>,
	e-mail <.mark@enron.com>, e-mail <.matt@enron.com>,
	e-mail <.mike@enron.com>, e-mail <.patricia@enron.com>,
	e-mail <.paul@enron.com>, e-mail <.randy@enron.com>,
	e-mail <.reagan@enron.com>, e-mail <.reid@enron.com>,
	e-mail <.richard@enron.com>, e-mail <.robert@enron.com>,
	e-mail <.scott@enron.com>, e-mail <.scott@enron.com>,
	e-mail <.scott@enron.com>, e-mail <.sean@enron.com>,
	e-mail <.shannon@enron.com>, e-mail <.travis@enron.com>,
	e-mail <.wayne@enron.com>, lloyd.will@enron.com, l..day@enron.com,
	jeff.king@enron.com, chris.dorland@enron.com"""
names = """Garcia, Miguel L. </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Mgarcia6>, 
           Seely, Michael </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Charles_Seely>, 
           Larkworthy, Carrie </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Clarkwor>, 
           Kinser, John </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Jkinser>, 
           Errigo, Joe </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Jerrigo>, 
           Ballato, Russell </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Rballato>, 
           Choate, Jason </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Jchoate>, 
           Hanse, Patrick </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Phanse>, 
           Burnett, Lisa </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Lburnett>, 
           David Nobles (E-mail) <david_nobles@ewingcc.com>, 
           Jeff Lenamon (E-mail) <jeff.lenamon@dzbank.de>, 
           Joe Ebert (E-mail) <jebert@eadslink.com>, 
           Joe Kuehler (E-mail) <jkuehler@wecon.com>, 
           Kari Tracey (E-mail) <KTracey111@aol.com>, 
           Lalena Stavinoha (E-mail) <lstavinoha@themedlehgroup.com>, 
           Mark Doggett (E-mail) <doggettmark@hotmail.com>, 
           Matt Doggett (E-mail) <matt_doggett@dell.com>, 
           Mike Newman (E-mail) <michael_j_newman@reliantenergy.com>, 
           Patricia Baughman (E-mail) <baughmanpatricia@hotmail.com>, 
           Paul  Henry (E-mail) <phenry@sempratrading.com>, 
           Randy Doggett (E-mail) <randy.doggett@pantellos.com>, 
           Reagan Marshall (E-mail) <themarshalls1@hotmail.com>, 
           Reid Stavinoha (E-mail) <reidstav@houston.rr.com>, 
           Richard Hrabal (E-mail) <richardhrabal@cs.com>, 
           Robert Pearson (E-mail) <rpearson@oilstates.com>, 
           Scott Sims (E-mail) <spsims@msn.com>, 
           Scott Smith (E-mail) <scottsmith@masterword.com>, 
           Scott Tichavsky (E-mail) <stichavsky@pct-inc.com>, 
           Sean Patrick Tracey (E-mail) <sean@traceylawfirm.com>, 
           Shannon Tyer (E-mail) <ShannonTyer@msn.net>, 
           Travis Baughman (E-mail) <travieso74@hotmail.com>, 
           Wayne Herndon (E-mail) <wayne.herndon@rweamericas.com>, 
           Will, Lloyd </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Lwill>, Day, Smith L. </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Sday>, King, Jeff </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Jking6>, Dorland, Chris </O=ENRON/OU=NA/CN=RECIPIENTS/CN=Cdorlan>"""

print(emailsAndNames(addresses, names))
```

```python
def nameFixes(name) :
    
    if (not isinstance(name, str)):
        name = ""

    regexp1 = re.compile('<[^>]+>')
    if regexp1.search(name) :
        name = re.sub('<[^>]+>', '', name)
        
    regexp2 = re.compile('\([^)]+\)')
    if regexp2.search(name) :
        name = re.sub('\([^)]+\)', '', name)
    
    regexp3 = re.compile('.+@.+')
    if regexp3.search(name) :
        name = re.sub('@.+', '', name)
        
    #regexp4 = re.compile('.+, .+')
    #if regexp4.search(name) :
    #    l = [x.strip() for x in name.split(',')]
    #    name = l[1] + ' ' + l[0]
    return name.strip()

def cleanEmails(s) :
    if (not isinstance(s, str)):
        s = ""
    
    pattern = re.compile(r'\s+')
    sentence = re.sub(pattern, '', s)
    return sentence.split(',')
```
