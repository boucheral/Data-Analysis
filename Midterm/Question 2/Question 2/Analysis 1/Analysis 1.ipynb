```python
def getLinks(path) :
    links = []
    names = glob.glob(path + '/*')
    
    for name in names :
        with open(name, 'r') as file :
            data = json.load(file)
            
        if data :
            links.extend(data.get('links', []))
            
    return list(set(links))
```
